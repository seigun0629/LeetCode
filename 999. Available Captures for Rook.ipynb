```python
class Solution:
        
    def numRookCaptures(self, board: List[List[str]]) -> int:
        rows = len(board)
        cols = len(board[0])
        res = None
        out = 0

        for i in range(rows):
            for j in range(cols):
                if board[i][j] == 'R':
                    res = (i,j)
        r,c = res
        #print(r,c,cols,rows)
        if res:
            while(c < cols):
                if board[r][c] == 'p':
                    out +=1
                    break

                elif board[r][c] == 'B':
                    # print(board[r][c+1])
                    break 

                c+=1
        r,c = res
        #print(r,c)
        if res:
            i = c-1
            while(i>= 0 ):
                if board[r][i] == 'p':
                    out +=1
                    break

                elif board[r][i] == 'B':
                    break 

                i-=1

        r,c = res
        #print(r,c)
        if res:
            i = r-1
            while(i >= 0 ):
                if board[i][c] == 'p':
                    out +=1
                    break

                elif board[i][c] == 'B':
                    break 

                i-=1

        r,c = res
       # print(r,c)
        if res:
            while(r < rows):
                if board[r][c] == 'p':
                    out +=1
                    break

                elif board[r][c] == 'B':
                    break 

                r+=1                                   

        return out
```
